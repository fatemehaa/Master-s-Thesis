```python
# specify data folder. The corresponding folder needs to be placed in the working directory.
# with wildcard for every file with a certain name
path = "PTB\\*.json"

# create empty dataframe
dfcontrol = pd.DataFrame()

qualitycheck = pd.DataFrame()

# read every file name in folder
for filename in glob.glob(path):
    with open(filename, 'r') as file:
        # make json files parsable
        data = "[" + file.read()
        data = data[:len(data)-2] + "]"

        # read data per file
        subjectdf = pd.read_json(data)

        # insert participant id in every line
        subjectdf.insert(0, "SubjectID", [int(
            filename[4:8])] * subjectdf.shape[0], True)
        qualitycheck = qualitycheck.append({"SubjectID": str(filename[4:8]), 'ExpStart': subjectdf.TimeStampBegin[0], "ExpEnd": subjectdf['TimeStampEnd'][55], "MissedTrials": len(
            subjectdf[subjectdf.TimeOut == True]), "LowestDeviation": min(subjectdf.Angle)}, ignore_index=True)
        dfcontrol = dfcontrol.append(subjectdf, ignore_index=True)
```

```python
def unnestN(dframe):
    # Un-nest nested data that is needed from the JSONs
    # Then each koordinate is inserted in a seperate column of the DataFrame
    
    PointerDirection = pd.DataFrame.from_records(
        dframe['PointerDirection'])
    dframe.insert(19, "PointerDirection_z",
                  PointerDirection['z'], True)
    dframe.insert(19, "PointerDirection_y",
                  PointerDirection['y'], True)
    dframe.insert(19, "PointerDirection_x",
                  PointerDirection['x'], True)
    
    PointerRotation = pd.DataFrame.from_records(
        dframe['PointerRotation'])
    dframe.insert(18, "PointerRotation_z",
                  PointerRotation['z'], True)
    dframe.insert(18, "PointerRotation_y",
                  PointerRotation['y'], True)
    dframe.insert(18, "PointerRotation_x",
                  PointerRotation['x'], True)
    
    PointerPosition = pd.DataFrame.from_records(
        dframe['PointerPosition'])
    dframe.insert(17, "PointerPosition_z",
                  PointerPosition['z'], True)
    dframe.insert(17, "PointerPosition_y",
                  PointerPosition['y'], True)
    dframe.insert(17, "PointerPosition_x",
                  PointerPosition['x'], True)

    ParticipantRotation = pd.DataFrame.from_records(
        dframe['ParticipantRotation'])
    dframe.insert(16, "ParticipantRotation_z",
                  ParticipantRotation['z'], True)
    dframe.insert(16, "ParticipantRotation_y",
                  ParticipantRotation['y'], True)
    dframe.insert(16, "ParticipantRotation_x",
                  ParticipantRotation['x'], True)

    
    ParticipantPosition = pd.DataFrame.from_records(
        dframe['ParticipantPosition'])
    dframe.insert(15, "ParticipantPosition_z",
                  ParticipantPosition['z'], True)
    dframe.insert(15, "ParticipantPosition_y",
                  ParticipantPosition['y'], True)
    dframe.insert(15, "ParticipantPosition_x",
                  ParticipantPosition['x'], True)
    
    ForcedRotation = pd.DataFrame.from_records(
        dframe['ForcedRotation'])
    dframe.insert(12, "ForcedRotation_z",
                  ForcedRotation['z'], True)
    dframe.insert(12, "ForcedRotation_y",
                  ForcedRotation['y'], True)
    dframe.insert(12, "ForcedRotation_x", 
                  ForcedRotation['x'], True)
   
    # Delete original columns from the data that has been unnested
    dframe = dframe.drop(['ForcedRotation',
                          'ParticipantPosition',
                          'ParticipantRotation',
                          'PointerPosition',
                          'PointerRotation',
                          'PointerDirection'], axis=1)
    return dframe


cNgroup = unnestN(dfNcontrol)
tNgroup = unnestN(dfNtest)
```
